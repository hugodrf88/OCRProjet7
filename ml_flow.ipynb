```python
X_train_trans=pd.read_csv("X_train_trans.csv")
X_test_trans=pd.read_csv("X_test_trans.csv")
y_train=pd.read_csv("y_train.csv",index_col=[0])
y_test=pd.read_csv("y_test.csv",index_col=[0])
```

```python
X_train_trans=X_train_trans.to_numpy()
X_test_trans=X_test_trans.to_numpy()
```

```python
from sklearn.metrics import confusion_matrix

# Définition de la fonction qui calcule le score métier personnalisé
def custom_scorer(y_true, y_pred):
    
    # Calcul de la matrice de confusion
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    # Calcul du score métier personnalisé
    score = (-10 * fn - fp + tn + 10 * tp) / (tn + tp + fn + fp)

    # Retourne le score métier personnalisé
    return score
```

```python
import mlflow
from sklearn.linear_model import LogisticRegression

mlflow.set_experiment("Log_reg")
mlflow.sklearn.autolog()
with mlflow.start_run(run_name="Log_reg_default"):
  logreg_model=LogisticRegression(C=170.06590074301735, 
                       penalty="l2",
                       solver="lbfgs",
                      max_iter=2000,)
  logreg_model.fit(X_train_trans,y_train)
  
  y_pred=logreg_model.predict(X_test_trans)
  metric_value=custom_scorer(y_pred,y_test)
  mlflow.log_metric("custom metric",metric_value)
```
